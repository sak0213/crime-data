```python
import pandas as pd
```

```python
#bring in crime CSVs
v_path = "../Source_Data/violent_crime_trimmed.csv"
d_path = "../Source_Data/drug_crime_trimmed.csv"
o_path = "../Source_Data/other_crime_trimmed.csv"
v_df=pd.read_csv(v_path)
d_df=pd.read_csv(d_path)
o_df=pd.read_csv(o_path)
```

```python
#combine them all together
all_crime_concat = pd.concat([v_df, d_df, o_df])
```

```python
all_crime_concat.describe()
```

```python
#trim unwanted columns
df = all_crime_concat.drop(['Unnamed: 0', 'year', 'fips_state_code', 'fips_county_code', 'fips_place_code', 'longitude', 'latitude'], axis=1)
```

```python
#group the data around fips codes
all_crime = df.groupby(['fips_state_county_code'], as_index=False).sum()
all_crime.describe()
```

```python
all_crime.to_csv('all_crime_merged.csv', index=False)
```
