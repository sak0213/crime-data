```python
import pandas as pd
```

```python
df = pd.read_csv("../Cleaned_Data/all_crime_merged.csv")
```

```python
df.columns
```

```python
#create a total crime column that's the sum of all columns minus fips and population
df["total_crime"] = df.iloc[:,2:].sum(axis=1)
df.head()
```

```python
#I'm impatient. I want to see the ML while i work
df_crime_adhoc = df[["fips_state_county_code", "total_crime"]]
df_county_data = pd.read_csv("county_data.csv")
df_all_adhoc = pd.merge(df_crime_adhoc, df_county_data, on="fips_state_county_code")
df_all_adhoc["total_crime_10k"] = df_all_adhoc["total_crime"]/df_all_adhoc["population_2020"]*10000
df_all_adhoc = df_all_adhoc.drop(columns=["total_crime"])
df_all_adhoc.to_csv("total_crime_adhoc.csv", index=False)
```
