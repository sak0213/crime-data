```python
from bs4 import BeautifulSoup as bs
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
url = 'https://livingwage.mit.edu/'
#visit url
browser.visit(url)
#parse HTML
html = browser.html
soup = bs(html, "html.parser")
browser.quit()
# Get lists of all the state URLs and Names
results = soup.find_all('li')
state_urls = []
state_names = []
for result in results:
    state_urls.append(result.find('a')['href'])
    state_names.append(result.find('a').text)
state_urls.pop(0)
state_urls.pop(0)
state_urls.pop(0)
state_urls.pop(0)
state_names.pop(0)
state_names.pop(0)
state_names.pop(0)
state_names.pop(0)
```

```python
df_living_wage = pd.read_html("https://livingwage.mit.edu/counties/42003")[0]
df_living_wage['State'] = 'Alabama'
df_living_wage['County'] = 'Autauga County'
df_living_wage = df_living_wage[0:0]
df_living_wage_factors = pd.read_html("https://livingwage.mit.edu/counties/42003")[1]
df_living_wage_factors['State'] = 'Alabama'
df_living_wage_factors['County'] = 'Autauga County'
df_living_wage_factors = df_living_wage_factors[0:0]
```

```python
def scrape_tables_lw(url, state, county):
    df_lw = pd.read_html(url)[0]
    df_lw['State'] = state
    df_lw['County'] = county
    df_lw["URL"] = url
    return df_lw
def scrape_tables_lw_factors(url, state, county):
    df_lwfacts = pd.read_html(url)[1]
    df_lwfacts['State'] = state
    df_lwfacts['County'] = county
    df_lwfacts["URL"] = url
    return df_lwfacts
    
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
for i in state_urls:
    url = 'https://livingwage.mit.edu' + i
    state = state_names[state_urls.index(i)]
    #visit url
    browser.visit(url)
    #parse HTML
    html = browser.html
    soup = bs(html, "html.parser")
    results = soup.find_all('li')
    county_urls = []
    county_names = []
    for result in results:
        county_urls.append(result.find('a')['href'])
        county_names.append(result.find('a').text)
    county_urls.pop(0)
    county_urls.pop(0)
    county_urls.pop(0)
    county_urls.pop(0)
    county_names.pop(0)
    county_names.pop(0)
    county_names.pop(0)
    county_names.pop(0)
    for i in county_urls:
        url_county = 'https://livingwage.mit.edu' + i
        county = county_names[county_urls.index(i)]
        df_living_wage =  df_living_wage.append(scrape_tables_lw(url_county, state, county))
        df_living_wage_factors = df_living_wage_factors.append(scrape_tables_lw_factors(url_county, state, county))
        print(state, county)
browser.quit()
    
    
```

```python
df_living_wage.to_csv('Source_Data/df_living_wage.csv')
df_living_wage_factors.to_csv('Source_data/df_living_wage_factors.csv')
```
