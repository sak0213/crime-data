```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Resources/ML_Ready_data/total_crime_adhoc.csv')
```

```python
df["total_crime_10k"].describe()
```

```python
mean = df["total_crime_10k"].mean()
stDev = df["total_crime_10k"].std()
```

```python
df["crime_lvl"]= df["total_crime_10k"].apply(lambda x: "Low" if ((x-mean)/stDev) < -0.5 else "High" if ((x-mean)/stDev) > 0.5 else "Medium")
```

```python
df_low = df[df["crime_lvl"] == "Low"]
df_medium = df[df["crime_lvl"] == "Medium"]
df_high = df[df["crime_lvl"] == "High"]
```

```python
#plot a histogram of each column in df
df_low.hist(bins=50, figsize=(20,20))
plt.show()
```

```python
#plot a histogram of each column in df
df_medium.hist(bins=50, figsize=(20,20))
plt.show()
```

```python
#plot a histogram of each column in df
df_high.hist(bins=50, figsize=(20,20))
plt.show()
```
